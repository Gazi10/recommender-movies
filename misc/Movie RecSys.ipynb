```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast','crew']]
```

movies.dropna(inplace=True)

```python
movies.duplicated().sum()
```

```python
def convert(dict):
    list = []
    for i in ast.literal_eval(dict):
        list.append(i['name'])
    return list
```

```python
import ast
```

```python
 movies['genres'] = movies['genres'].apply(convert)
```

```python
 movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convertCast(dict):
    list = []
    count = 0
    for i in ast.literal_eval(dict):
        if count != 3:
            list.append(i['name'])
            count+=1
        else:
            break
    return list
```

```python
movies['cast'] = movies['cast'].apply(convertCast)
```

```python
def fetchDirec(dict):
    list = []
    for i in ast.literal_eval(dict):
        if i['job'] == 'Director':
            list.append(i['name'])
            break
    return list
```

```python
movies['crew'] = movies['crew'].apply(fetchDirec)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split() if type(x) == str else x)
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ', '') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ', '') for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview']+ movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] 
```

```python
movies.head()
```

```python
dataFrame = movies[['movie_id', 'title', 'tags']]
```

```python
dataFrame
```

```python
dataFrame['tags'] = dataFrame['tags'].apply(lambda x:" ".join(x) if type(x) == list else x)
```

```python
dataFrame['tags'] = dataFrame['tags'].apply(lambda x:x.lower() if type(x) == str else x)
```

```python
dataFrame.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(dataFrame['tags'].values.astype('U')).toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    list = []
    
    if type(text) == str:
        for i in text.split():
            list.append(ps.stem(text))
    return " ".join(list)
```

```python
dataFrame['tags'] = dataFrame['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movieIndex = dataFrame[dataFrame['title'] == movie].index[0]
    distances = similarity[movieIndex]
    moviesList = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])
    
    for i in moviesList:
        print(dataFrame.iloc[i[0]].title)
```

```python
recommend('Superman')
```
